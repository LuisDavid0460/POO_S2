### BREAK

**break:** sale del bucle
inmediatemente, e incondicionalmente

```python
## Agregamos una condicion para que se salga del ciclo cuando se cumpla

i = 1
while i <= 1000:
  print(i)
  if i == 10:
    print("Entro al if")
    break
  i += 1
print("Fuera del while")
```

```python
while(True):
  print("""Selecciona una opcion
  1) Opcion 1
  2) Opcion 2
  3) Salir""")
  opcion = input("多Cual es la opcion que deseas? ")
  if opcion == "1":
    print("Seleccionaste la Opcion 1")
  elif opcion == "2":
    print("Seleccionaste la Opcion 2")
  elif opcion == "3":
    print("Seleccionaste la Opcion Salir")
    break
  else:
    print("Opcion desconocida, vuelve a intentarlo")
  print("Fuera del ciclo")
```
